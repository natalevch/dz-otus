# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
tips['total_bill'].plot(
    kind='hist', # тип графика: гистограмма
    title='Распределение признака total_bill'
)
plt.xlabel('Value, $');
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
tips.plot(
    kind='scatter', # тип графика: диаграмма рассеяния
    x='tip',      # переменная для оси X
    y='total_bill',       # переменная для оси Y
    alpha=0.5,      # alpha - прозрачность точек, помогает увидеть их плотность
    title='Зависимость чаевых от счета'
);
```

### Постройте pairplot

```python
sns.pairplot(
    tips,
    hue='total_bill' # зависимости отtotal_bill
);
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
tips.plot(
    kind='scatter', # тип графика: диаграмма рассеяния
    x='day',      # переменная для оси X
    y='total_bill',       # переменная для оси Y
    alpha=0.5,      # alpha - прозрачность точек, помогает увидеть их плотность
    title='Зависимость общей суммы счета от дня недели'
);
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
sns.histplot(
    data=tips,
    x='tip',
    hue = 'time',
    kde=True,
    palette=['steelblue', 'lightblue']
)
plt.title('распределение признака tip в зависимости от категорий time');
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# График 1: Мужчины
sns.scatterplot(
    data=tips[tips['sex']=='Male'],
    x='total_bill', # Ось X: общий счёт
    y='tip', # Ось Y: чаевые
    hue='smoker', # Цвет по признаку smoker
    alpha=0.6 # Прозрачность точек
)
plt.title('Чаевые мужчин по общему счёту (Male)');

```

```python
# График 2: Женщины
sns.scatterplot(
    data=tips[tips['sex']=='Female'], # Только женщины
    x='total_bill', # Ось X: общий счёт
    y='tip', # Ось Y: чаевые
    hue='smoker', # Цвет по по признаку smoker
    alpha=0.6 # Прозрачность точек
)
plt.title('Чаевые женщин по общему счёту (Female)');
```

## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

```python
corr_matrix = tips.select_dtypes(include=np.number).corr()
sns.heatmap(
    corr_matrix,
    annot=True,     # annot=True - подписать значения в ячейках
    cmap='viridis', # cmap - цветовая карта
    fmt='.2f'       # fmt='.2f' - формат чисел (2 знака после запятой)
)
plt.title('Матрица корреляций признаков');
```

Выводы:
1.  Мужчины оставляют более крупные чаевые при больших счетах, чем женщины (график scatterplot)
2. Гистограммы tip по time показывают, что за ужин (Dinner) чаевые значительно выше, чем за обед (Lunch)
3. График показывает, что счет больше на выходных (Sat, Sun), это коррелирует с повышенными чаевыми
